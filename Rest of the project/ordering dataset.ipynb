```python
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
l=glob.glob('./database/*.jpg')
img=[cv2.imread(file,0) for file in glob.glob('./database/*.jpg')]
```

```python
d={}
D={}
for i in range(len(l)):
    p=int(l[i][13:16])-1
    if p not in d.keys():
        d[p]=[]
        d[p].append(l[i])
        D[p]=[]
        D[p].append(img[i])
    else:
        d[p].append(l[i])
        D[p].append(img[i])
    
name_l=[]    
img_l=[]
for key, value in d.items():
    name_l.append(value)
    
for key, value in D.items():
    value=np.asarray(value)
    img_l.append(value)
```

```python
name_L=[]
img_L=[]
for i in range(len(name_l)):
    templ=[l for l in name_l[i] if l[16]=='L']
    tempr=[l for l in name_l[i] if l[16]=='R']
    name_L.append([templ,tempr])
    img_L.append([img_l[i][:len(templ),:],img_l[i][len(templ):,:]])
```
